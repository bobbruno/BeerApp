# Generating User Stem Data

````python
import psycopg2
import pandas as pd

with psycopg2.connect(database='BeerDb', user='postgres', password='postgres', host='localhost', port=5432) as conn:
    df = pd.read_sql('select * from tb_beer_features', conn)
````

````python
from joblib import Parallel, delayed

def readDb(minUser, maxUser):
    with psycopg2.connect(database='BeerDb', user='postgres', password='postgres', host='localhost', port=5432) as conn:
        sql = '''select r.user_id, s.word_stem, 
        sum((s.word_stem = any (r.rating_words))::int)::float / count(*)  as stemPercent
        from tb_stem_dict s,
             (select t1.* from tb_rating_notes t1,
                               tb_rating t2
                          where t1.user_id between %s and %s and
                                t1.beer_id = t2."BEER_ID" and
                                t1.user_id = t2."USER_ID" and
                                t1.rating_date = t2."RATING_DATE" and
                                t2."RATING_COMPOUND" > 3.5) as r
        where s.is_keyword
        group by 1, 2
        order by 1, 2'''
        return pd.read_sql_query(sql, conn, coerce_float=True, 
                                 params=(minUser, maxUser))
nBlocks = 30
nUsers = 337432
blockSize = nUsers // nBlocks
dfList = Parallel(n_jobs = 13)(delayed(readDb)(i*blockSize, (i+1)*blockSize-1) for i in xrange(nBlocks+1))
dfUser = pd.concat(dfList)
x=dfUser.pivot(index='user_id', columns='word_stem', values='stempercent')
````

````python
with psycopg2.connect(database='BeerDb', user='postgres', password='postgres', host='localhost', port=5432) as conn:
    cur = conn.cursor()
    try:
        sql = u'CREATE TABLE TB_USER_STEM_PERCENT (USER_ID integer not null,' +\
              ' double precision,'.join(x.columns) + ' double precision)'
        cur.execute(sql)
    except:
        conn.commit()
        sql = 'truncate table TB_USER_STEM_PERCENT'
        cur.execute(sql)
        conn.commit()
    blockSize = 10000
    record = []
    sql = u'%s,'*207
    sql = u'insert into TB_USER_STEM_PERCENT values(' + sql
    sql =  sql[:-1] + u')'
    for i in x.index.values:
        record.append((i,)+ tuple(x.loc[i]))
        if i and not i % blockSize:
            #print sql
            cur.executemany(sql, record)
            record = []
            conn.commit()
    if len(record):
        cur.executemany(sql, record)
        record = []
        conn.commit()
````

````python
if 'cluster' in dfBFeatures.columns.tolist():
    dfTemp = dfBFeatures.copy().drop('cluster', axis=1)
else:
    dfTemp = dfBFeatures
````
